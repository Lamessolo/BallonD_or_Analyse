```python
import requests
url ="https://fr.wikipedia.org/wiki/Ballon_d%27or_2022"
response = requests.get(url)
response = response.content
response
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response,"html.parser")
```

```python
thead =soup.find('thead')
thead
```

```python
soup
```

```python
#retrieve Table balise 
table =soup.find('table',class_="wikitable alternance sortable")
table
```

```python
tbody = table.find('tbody')
tbody
```

```python
for joueur in tbody :
    lignejoueur = tbody.find_all('tr')
    print (lignejoueur)
```

```python
for lignetd in lignejoueur:
        lignejoueurtd = lignetd.find_all('td')
        print(lignejoueurtd)
```

```python
# Retrieve la liste des joueurs et leur differents attribut
joueur_rows =[]
for joueur in tbody.find_all('tr') :
    for y in joueur.find_all('td'):
        #print(y.text)
        joueur_rows.append(y.text)
        #for a in y.find_all('a'):
            #print(a.text)
joueur_rows
```

```python
for playerretrieve in joueur_rows:
    print(playerretrieve)
```
